```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import sklearn.metrics
```

Read the txt file where the anomaly score is writen

```python
data = pd.read_csv("number_results.txt", header=None)
data1 = pd.read_csv("number_results2.txt", header=None)
data2 = pd.read_csv("number_results9.txt", header=None)
```

Label the data (English = 0, Other = 1)

```python
def labeling(data_list):
    label = np.zeros(124)
    for x in range(0,len(data_list)-124):
        label = np.append(label, 1)
    return label
```

```python
data[1] = labeling(data)
data1[1] = labeling(data1)
data2[1] = labeling(data2)
```

Sort the anomaly according to their score from min to max

```python
def sorting (data):
    sorted_probability, sorted_label = zip(*sorted(zip(data[0], data[1])))
    tagalog_one = np.sum(sorted_label)
    english_one = len(sorted_label) - tagalog_one
    sorted_lb = np.array(sorted_label)
    sensitivity, specificity = sen_spe(sorted_lb, english_one, tagalog_one)
    return sensitivity, specificity
```

Calculate the sensitivity and specificity

```python
def sen_spe(sorted_lb, english_one, tagalog_one):
    sensitivity = []
    specificity = []
    for x in range(0, len(sorted_lb)):
        sumofOne = sorted_lb[x:-1].sum()
        sensitivity.append(sumofOne/tagalog_one)
        specificity.append((len(sorted_label)-sumofOne-x)/english_one)
    return sensitivity, specificity
```

```python
sensitivity1,specificity1 = sorting (data)
```

Plot the results of the sensitivity and specificity

```python
pt.plot(specificity1, sensitivity1)
pt.show()

print(sensitivity1)
print(specificity1)
```

Calculate the AUC

```python
sklearn.metrics.auc(specificity1, sensitivity1)
```
