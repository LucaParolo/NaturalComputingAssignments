```python
import matplotlib.pyplot as plt
import csv
```

```python
import numpy as np
```

```python
best_fitness=[]
mean_fitness=[]
with open('fitness_history.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        best_fitness.append(float(row[0]))
        mean_fitness.append(float(row[1]))
```

```python
first_game_best=[0,]
second_game_best=[]

for i in range(1,len(best_fitness),2):
    first_game_best.append(best_fitness[i])
    second_game_best.append(best_fitness[i+1])
    
```

```python
first_game_mean=[0,]
second_game_mean=[]

for i in range(1,len(mean_fitness),2):
    first_game_mean.append(mean_fitness[i])
    second_game_mean.append(mean_fitness[i+1])
    
```

```python
first_game_mean
```

```python
first_game_mean=(first_game_mean-np.min(first_game_mean))/(np.max(first_game_mean)-np.min(first_game_mean))
```

```python
first_game_mean
```

```python
second_game_mean=(second_game_mean-np.min(second_game_mean))/(np.max(second_game_mean)-np.min(second_game_mean))
```

```python
first_game_best=(first_game_best-np.min(first_game_best))/(np.max(first_game_best)-np.min(first_game_best))
```

```python
second_game_best=(second_game_best-np.min(second_game_best))/(np.max(second_game_best)-np.min(second_game_best))
```

```python
first_game_best=np.delete(first_game_best,0)
```

```python
final_best=[0,]
for i in range(1,len(first_game_best)):
    final_best.append(first_game_best[i])
    final_best.append(second_game_best[i])
```

```python
first_game_mean=np.delete(first_game_mean,0)
```

```python
final_mean=[0,]
for i in range(1,len(first_game_mean)):
    final_mean.append(first_game_mean[i])
    final_mean.append(second_game_mean[i])
```

```python
plt.title("Normalized mean score for multiple environment")
plt.plot(first_game_mean,label="Space Invaders")
plt.plot(second_game_mean,label="Demon Attack")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("2gamesmean")
plt.show()
```

```python
plt.title("Normalized best score for multiple environment")
plt.plot(first_game_best,label="Space Invaders")
plt.plot(second_game_best,label="Demon Attack")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("2gamesbest")
plt.show()
```

```python
plt.title("Normalized best score for multiple environment ")
plt.plot(final_best,label="Both games")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("2gamesbestfinal")
plt.show()
```

```python
plt.title("Normalized mean score for multiple environment ")
plt.plot(final_mean,label="Both games")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("2gamesmeanfinal")
plt.show()
```
