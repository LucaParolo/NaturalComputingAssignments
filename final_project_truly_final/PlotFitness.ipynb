```python
import matplotlib.pyplot as plt

import csv
import numpy
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 14
BIGGER_SIZE = 18

      # controls default text sizes
plt.rcParams['font.weight'] = 'normal'
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
```

```python
best_fitness_1=[]
mean_fitness_1=[]
with open('fitness1.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        best_fitness_1.append(float(row[0]))
        mean_fitness_1.append(float(row[1]))
best_fitness_2=[]
mean_fitness_2=[]
with open('fitness2.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        best_fitness_1.append(float(row[0]))
        mean_fitness_1.append(float(row[1]))
```

```python
best_fitness=np.concatenate(best_fitness_1,best_fitness_2)
```

```python
fitness_1=[]
fitness_2=[]
fitness_3=[]
fitness_4=[]
with open('species_fitness.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        for i in range(len(row)):
            if row[i]=='NA':
                row[i]=0
        fitness_1.append(float(row[0]))
        fitness_2.append(float(row[1]))
        fitness_3.append(float(row[2]))
        fitness_4.append(float(row[3]))
```

```python
plt.title("Fitness for SpaceInvaders")
plt.plot(best_fitness,label="best fitness")
plt.plot(mean_fitness,label="mean fitness")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("FitnessSpaceInvaders")
plt.show()
```

```python
plt.title("Fitness for SpaceInvaders by Species")
plt.plot(fitness_1,label="Fitness species 1")
plt.plot(fitness_2,label="Fitness species 2")
plt.plot(fitness_3,label="Fitness species 3")

plt.plot(fitness_4,label="Fitness species 4")

plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("FitnessSpaceInvadersSpecies")
plt.show()
```

```python
pop_1=[]
pop_2=[]
pop_3=[]
pop_4=[]
with open('speciation.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        for i in range(len(row)):
            if row[i]=='NA':
                row[i]=0
        pop_1.append(float(row[0]))
        pop_2.append(float(row[1]))
        pop_3.append(float(row[2]))
        pop_4.append(float(row[3]))
```

```python
plt.title("Population for SpaceInvaders by Species")
plt.plot(pop_1,label="Population species 1")
plt.plot(pop_2,label="Population species 2")
plt.plot(pop_3,label="Population species 3")

plt.plot(pop_4,label="Population species 4")

plt.legend()
plt.xlabel("Generation")
plt.ylabel("# of Individuals")
plt.savefig("PopulationSpaceInvaders")
plt.show()
```
