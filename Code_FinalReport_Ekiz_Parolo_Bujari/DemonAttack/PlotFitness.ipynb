```python
import matplotlib.pyplot as plt

import csv
import numpy as np
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 14
BIGGER_SIZE = 18

      # controls default text sizes
plt.rcParams['font.weight'] = 'normal'
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
```

```python
best_fitness=[]
mean_fitness=[]
with open('fitness_history.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        best_fitness.append(float(row[0]))
        mean_fitness.append(float(row[1]))
```

```python
both=np.concatenate((best_fitness,mean_fitness),axis=0)
```

```python
mean_fitness=(mean_fitness-np.min(both))/(np.max(both)-np.min(both))
best_fitness=(best_fitness-np.min(both))/(np.max(both)-np.min(both))
```

```python
plt.title("Normalized Fitness for DemonAttack")
plt.plot(best_fitness,label="best fitness")
plt.plot(mean_fitness,label="mean fitness")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.savefig("FitnessDemonAttack")
plt.show()
```
